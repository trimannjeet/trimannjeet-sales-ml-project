```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create target (classification)
data["Sales_Category"] = data["Sales"].apply(lambda x: 1 if x > 200 else 0)

# Features & target
X = data.drop(["Sales", "Sales_Category"], axis=1)
y = data["Sales_Category"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Accuracy
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd

# Load data
data = pd.read_csv("train.csv")

# Drop unnecessary columns safely
drop_cols = ["Row ID", "Order ID", "Customer ID", "Customer Name", "Product ID", "Product Name"]
data = data.drop(columns=[col for col in drop_cols if col in data.columns])

# Remove missing values
data = data.dropna()

data = data.drop(columns=["Order Date", "Ship Date"])

# Convert categorical → numeric
data = pd.get_dummies(data, drop_first=True)

# ✅ Define X and y (REGRESSION)
X = data.drop("Sales", axis=1)
y = data["Sales"]

# Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, y_pred))
```
